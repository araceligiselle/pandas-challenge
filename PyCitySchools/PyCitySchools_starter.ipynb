```python
# Dependencies and Setup
import pandas as pd
import numpy as np
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## Local Government Area Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average maths score 

* Calculate the average reading score

* Calculate the percentage of students with a passing maths score (50 or greater)

* Calculate the percentage of students with a passing reading score (50 or greater)

* Calculate the percentage of students who passed maths **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Total number of schools
total_schools = len(school_data_complete["school_name"].unique())

# Total number of students
total_students = school_data_complete["Student ID"].count()

# Total Budget
total_budget = school_data["budget"].sum()

# Average Math Score
average_math = school_data_complete["maths_score"].mean()

# Average Reading Score
average_reading = school_data_complete["reading_score"].mean()

# Maths Passing score
passing_math = school_data_complete[(school_data_complete["maths_score"]>=50)].count()["Student ID"]
passing_percentage_math = (passing_math / total_students) * 100

# Reading Passing score
passing_reading = school_data_complete[(school_data_complete["reading_score"]>=50)].count()["Student ID"]
passing_percentage_reading = (passing_reading / total_students) * 100 

# Maths + Reading Passing score
overall_passing = school_data_complete[((school_data_complete["maths_score"] >= 50) & (school_data_complete["reading_score"] >= 50))].count()["Student ID"]
overall_percentage = (overall_passing/total_students) * 100

# DataFrame
area_summary= [{"Total Schools": total_schools,
                 "Total Students": "{:,}".format(total_students),
                 "Total Budget": "${:,.2f}".format(total_budget),
                 "Average Maths Score": average_math,
                 "Average Reading Score": average_reading,
                 "% Passing Maths": passing_percentage_math,
                 "% Passing Reading": passing_percentage_reading,
                 "% Overall Passing": overall_percentage}]
area_summary_df = pd.DataFrame(area_summary)
area_summary_df.head()
```

## School Summary

* Create an overview table that summarises key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed maths **and** reading.)
  
* Create a dataframe to hold the above results

```python
#School name and type
school_overview = school_data_complete.groupby("school_name")["type"].first()

#Total students
students_overview = school_data_complete["school_name"].value_counts()

#Total school budget
total_school_budget = school_data_complete.groupby("school_name")["budget"].mean()

#Per student budget
per_student_budget = total_school_budget / students_overview

#Average Maths score
average_math_overview = school_data_complete.groupby("school_name")["maths_score"].mean()

#Average reading score
average_reading_overview = school_data_complete.groupby("school_name")["reading_score"].mean()

#% Passing Maths
passing_math_overview = school_data_complete[school_data_complete["maths_score"]>=50].groupby("school_name")["Student ID"].count() / students_overview * 100

#% Passing Reading
passing_reading_overview = school_data_complete[school_data_complete["reading_score"]>=50].groupby("school_name")["Student ID"].count() / students_overview * 100

#% Overall Passing
overall_passing_overview = school_data_complete[((school_data_complete["maths_score"] >= 50) & (school_data_complete["reading_score"] >= 50))].groupby("school_name")["Student ID"].count() / students_overview *100

#Data Frame Overview
per_school_summary = {"School Type": school_overview,
                 "Total Students": students_overview,
                 "Total School Budget": total_school_budget.map("${:,.2f}".format),
                 "Per Student Budget": per_student_budget.map("${:,.2f}".format),
                 "Average Maths Score": average_math_overview,
                 "Average Reading Score": average_reading_overview,
                 "% Passing Maths": passing_math_overview,
                 "% Passing Reading": passing_reading_overview,
                 "% Overall Passing": overall_passing_overview}
per_school_summary_df = pd.DataFrame(per_school_summary)
per_school_summary_df
```

## Maths Scores by Year

  * Create a pandas series for each year. Hint: use a conditional statement.
  
  * Group each series by school

```python
#Calculate maths's mean per grade
maths_score_9 = school_data_complete[school_data_complete["year"] == 9]
maths_score_9_average = maths_score_9.groupby(["school_name"])["maths_score"].mean()

maths_score_10 = school_data_complete[school_data_complete["year"] == 10]
maths_score_10_average = maths_score_10.groupby(["school_name"])["maths_score"].mean()

maths_score_11 = school_data_complete[school_data_complete["year"] == 11]
maths_score_11_average = maths_score_11.groupby(["school_name"])["maths_score"].mean()

maths_score_12 = school_data_complete[school_data_complete["year"] == 12]
maths_score_12_average = maths_score_12.groupby(["school_name"])["maths_score"].mean()

#Data Frame overview
maths_score_by_year = pd.DataFrame({"Year 9": maths_score_9_average,
    "Year 10": maths_score_10_average,
    "Year 11": maths_score_11_average,
    "Year 12": maths_score_12_average})

maths_score_by_year
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]

school_performance = per_school_summary_df.copy()
per_student_budget = pd.to_numeric(per_student_budget)

#Slice the data and placed it into bins
school_performance["Spending Ranges"] = pd.cut(per_student_budget, spending_bins, labels=labels)

performance_by_spending = school_performance.groupby(["Spending Ranges"])

average_math_overview = performance_by_spending["Average Maths Score"].mean()
average_reading_overview = performance_by_spending["Average Reading Score"].mean()
passing_math_overview = performance_by_spending["% Passing Maths"].mean()
passing_reading_overview = performance_by_spending["% Passing Reading"].mean()
overall_passing_overview = performance_by_spending["% Overall Passing"].mean()

#Data Frame
spending_summary = pd.DataFrame({
                "Average Maths Score": average_math_overview.map("{:,.2f}".format),
                "Average Reading Score": average_reading_overview.map("{:,.2f}".format),
                "% Passing Maths": passing_math_overview.map("{:,.2f}".format),
                "% Passing Reading": passing_reading_overview.map("{:,.2f}".format),
                 "% Overall Passing": overall_passing_overview.map("{:,.2f}".format)})
spending_summary
```

* Perform the same operations as above, based on school size.

```python
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

#Slice the data and placed it into bins
per_school_summary_df["School Size"] = pd.cut(per_school_summary_df["Total Students"], size_bins, labels=labels)

average_math_overview = per_school_summary_df.groupby(["School Size"])["Average Maths Score"].mean()
average_reading_overview = per_school_summary_df.groupby(["School Size"])["Average Reading Score"].mean()
passing_math_overview = per_school_summary_df.groupby(["School Size"])["% Passing Maths"].mean()
passing_reading_overview = per_school_summary_df.groupby(["School Size"])["% Passing Reading"].mean()
overall_passing_overview = per_school_summary_df.groupby(["School Size"])["% Overall Passing"].mean()

#Data Frame
size_summary = pd.DataFrame({
                "Average Maths Score": average_math_overview.map("{:,.2f}".format),
                "Average Reading Score": average_reading_overview.map("{:,.2f}".format),
                "% Passing Maths": passing_math_overview.map("{:,.2f}".format),
                "% Passing Reading": passing_reading_overview.map("{:,.2f}".format),
                 "% Overall Passing": overall_passing_overview.map("{:,.2f}".format)})
size_summary
```
